```python
a = np.array([i for i in range(100000)])
b = np.array([i for i in range(100000)])
```

```python
import time

class Timer:
    def __init__(self, name):
        self.name = name
    def __enter__(self):
        self._enter_time = time.time()

    def __exit__(self, *exc_args):
        self._exit_time = time.time()
        print(f"{self.name} = {1000*(self._exit_time - self._enter_time):.2f}ms")
```

```python
tic = time.time()
dot = 0.0
for i in range(len(a)):
    dot += a[i] * b[i]
toc = time.time()

print("dot_product = "+str(dot))
print("Computation time = " + str(1000*(toc-tic)) + "ms")
```

```python
n_tic = time.time()
n_dot_product = np.array(a).dot(np.array(b))
n_toc = time.time()

print("\nn_dot_product = "+str(n_dot_product))
print("Computation time = " + str(1000*(n_toc-n_tic)) + "ms")
```

```python
def power_iteration(A, tolerance=1e-7):
    b_old = np.random.rand(A.shape[1])
    b = np.random.rand(A.shape[1])
    num_iterations = 0
    while num_iterations == 0 or np.linalg.norm(b_old - b) > tolerance:
        b_old = np.copy(b)
        b = np.dot(A, b)
        b_norm = np.linalg.norm(b)
        b /= b_norm
        num_iterations += 1
    return np.dot(A, b), b, num_iterations
```

```python
A = np.array([[1, 2], [2, 1]])
ab, b, number_iterations = power_iteration(A, 1e-12)

eig1 = ab[0] / b[0]
eig2 = ab[1] / b[1]
assert(np.abs((eig1 - eig2)) < 1e-5)

#b /= b[1]

print(eig1, b, number_iterations)

print(np.linalg.norm([3, 6]))
```

#   

$$
A \in (\mathbb{R})_{mxn} \\
B \in (\mathbb{R})_{nxk} \\
A.B \in (\mathbb{R})_{mxk}
$$
$$
(A . B)_{i, j} = \sum_{t=0}^{n-1} A_{i,t} B_{t, j}
$$

```python
# solution
def matmul(A, B):
    assert A.shape[1] == B.shape[0]
    m = A.shape[0]
    n = A.shape[1]
    k = B.shape[1]
    
    result = np.zeros((m, k))
    for i in range(m):
        for j in range(k):
            for t in range(n):
                result[i][j] += A[i][t] * B[t][j]
    return result
```

```python
# solution - alternative
Amat = np.random.rand(70, 100)
Bmat = np.random.rand(100, 101)

with Timer("my matmul"):
    ans1 = matmul(Amat, Bmat)

with Timer("numpy matmul"):
    ans2 = np.matmul(Amat, Bmat)

assert np.allclose(ans1, ans2)
```

```python
# solution - alternative
def matmulFaster(A, B):
    assert A.shape[1] == B.shape[0]
    m = A.shape[0]
    n = A.shape[1]
    k = B.shape[1]
    
    result = np.zeros((m, k))
    for i in range(m):
        for j in range(k):
            result[i][j] = np.dot(A[i], np.transpose(B)[j])
    return result
```

```python
with Timer("faster matmul"):
    ans3 = matmulFaster(Amat, Bmat)
```
